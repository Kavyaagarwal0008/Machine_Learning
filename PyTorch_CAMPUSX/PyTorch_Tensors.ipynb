```python
import time

size=10000

matrix_cpu1=torch.randn(size,size)
matrix_cpu2=torch.randn(size,size)


#time on cpu
start_time=time.time()
result_cpu=torch.matmul(matrix_cpu1,matrix_cpu2)
cpu_time=time.time()-start_time
print(f"Time taken by cpu {cpu_time}")

#move matrices to GPU
matrix_cpu1=matrix_cpu1.to('cuda')
matrix_cpu2=matrix_cpu2.to('cuda')

#Measure time on GPU
start_time=time.time()
result_gpu=torch.matmul(matrix_cpu1,matrix_cpu2)
torch.cuda.synchronize()
gpu_time=time.time()-start_time
print(f"Time taken by gpu {gpu_time}")

print("\nSpeedup (CPU /GPU TIME ):", cpu_time/gpu_time)
```
